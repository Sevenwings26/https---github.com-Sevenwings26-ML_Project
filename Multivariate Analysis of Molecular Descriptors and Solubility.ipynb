# Machine Learning Project

## Load Dataset  
**About the Dataset:**  
This dataset contains information on whether a molecule is soluble in water or a solvent, based on its descriptors.

**Source:**  
[Delaney Solubility Dataset](https://github.com/dataprofessor/data/blob/master/delaney_solubility_with_descriptors.csv)

---

**Multivariate Analysis:**  
Next, we will perform multivariate analysis on this dataset to explore the relationships between various molecular descriptors and their solubility.  

```python
import pandas as pd

df = pd.read_csv("/content/delaney_solubility_with_descriptors.csv")
df.head()
```

```python
# MolLogP	MolWt	NumRotatableBonds	AromaticProportion are X and dependent on logS as Y
df.shape
```

```python
df.info()
# df.describe()
```

## Data preparation

### data sepaartion, x and y

```python
y = df['logS']
y
```

```python
x = df.drop("logS", axis=1)
x.head(5)
```

### Data splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
# x_train # 80% of dataset
# x_test # 20% of dataset
```

## Model Building

### Linear Regression

#### Training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

#### Applying the model to make predcition

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

#### Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_mean = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mean = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print("LR MSE (Train):", lr_train_mean)
print(f"LR R2 (Train):, {lr_train_r2} \n")
print("LR MSE (Test):", lr_test_mean)
print("LR R2 (Test):", lr_test_r2)
```

```python
# display in tabular form
lr_results = pd.DataFrame(['Linear regression', lr_train_mean, lr_train_r2, lr_test_mean, lr_test_r2]).transpose()
lr_results.columns = ["Method", "Training MSE", "Training R2", "Test MSE", "Test R2"]
```

```python
lr_results
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

#### Applying the model to make predcition

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

#### Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mean = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mean = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
# display in tabular form
rf_results = pd.DataFrame(['Random Forest', rf_train_mean, rf_train_r2, rf_test_mean, rf_test_r2]).transpose()
rf_results.columns = ["Method", "Training MSE", "Training R2", "Test MSE", "Test R2"]
```

```python
rf_results
```

## Model comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
```
